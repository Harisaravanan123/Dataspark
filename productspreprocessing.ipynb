```python
nullvalues=df['Unit Price USD'].isna().sum()
print(nullvalues)
```

```python
df['Unit Price USD'] = (
    df['Unit Price USD']
    .str.replace(r'\$', '', regex=True)    # Remove dollar sign
    .str.replace(',', '', regex=False)      # Remove commas
    .str.strip()                            # Remove any extra spaces
    .astype(float)                          # Convert to float
)
```

```python
df['Unit Cost USD'] = (
    df['Unit Cost USD']
    .str.replace(r'\$', '', regex=True)    # Remove dollar sign
    .str.replace(',', '', regex=False)      # Remove commas
    .str.strip()                            # Remove any extra spaces
    .astype(float)                          # Convert to float
)
```

```python
#To create a database
import psycopg2
from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT

connection = psycopg2.connect(
    host="localhost",
    port="5432",
    database="Dataspark",
    user="postgres",
    password="12345678"
)

connection.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)

writer = connection.cursor()

#writer.execute("create database hello")

writer.execute("""create table products("ProductKey" integer,"Product Name" text,"Brand" text,"Color" text,
               "Unit Cost USD" real,"Unit Price USD" real,"SubcategoryKey" integer,"Subcategory" text,"CategoryKey" integer,"Category" text)""")

writer.close()
connection.close()
```
