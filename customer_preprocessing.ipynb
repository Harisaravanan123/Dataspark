```python
#To create a database
import psycopg2
from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT

connection = psycopg2.connect(
    host="localhost",
    port="5432",
    database="Dataspark",
    user="postgres",
    password="12345678"
)

connection.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)

writer = connection.cursor()

#writer.execute("create database hello")

writer.execute("""create table customer ("CustomerKey" INTEGER,"Gender" text,"Name" text,"City" text,"State Code" text,
               " State" text,"Zip Code" text,"Country" text," Continent" text,"Birthday" date)""")

writer.close()
connection.close()
```
