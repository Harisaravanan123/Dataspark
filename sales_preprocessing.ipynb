```python
import psycopg2

# Establishing the connection
connection = psycopg2.connect(
    host="localhost",
    database="Dataspark",
    user="postgres",
    password="12345678"  # Password should be in quotes
)

# Creating a cursor object
cursor = connection.cursor()

# Executing a sample query
cursor.execute("SELECT version()")

# Fetching the result
version = cursor.fetchone()

# Printing the PostgreSQL version
print("You are connected to - ", version)

# Closing the connection
connection.close()
```

```python
#To create a database
import psycopg2
from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT

connection = psycopg2.connect(
    host="localhost",
    port="5432",
    database="Dataspark",
    user="postgres",
    password="12345678"
)

connection.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)

writer = connection.cursor()

#writer.execute("create database hello")

writer.execute("""create table sales2 ("Order Number" INTEGER,"Line Item" INTEGER,"Order Date" DATE,"CustomerKey" INTEGER,"StoreKey" INTEGER,"ProductKey" INTEGER,"Quantity" INTEGER,"Currency Code" TEXT)""")

writer.close()
connection.close()
```

```python
import pandas as pd
from sqlalchemy import create_engine

# pip install pymysql
# import pymysql


# Replace with your connection details
host="localhost"
port="5432"
database="Dataspark"
user="postgres"
password="12345678"

# Create the connection string (URL format) -> postgres
engine_string = f"postgresql://{"postgres"}:{12345678}@{"localhost"}:{5432}/{"Dataspark"}"

# Create the connection string (URL format) -> mysql
# engine_string = f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"

# Create the SQLAlchemy engine
engine = create_engine(engine_string)

# Sample DataFrame (replace with your actual data)

df = pd.DataFrame(df)

# Table name where you want to push the data
table_name = "sales2"

# Push the DataFrame to the PostgreSQL table
df.to_sql(table_name, engine,if_exists='replace', index=False)  #['fail', 'replace', 'append']

print("Data successfully pushed to PostgreSQL table!")
```
